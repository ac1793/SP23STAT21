## This is another Markdown cell

When you work with Jupyter notebooks you are simply moving back and forth between Markdown cells and Code cells.  You can also create graphics within a code cell (you'll learn all of this eventually):

```python
import matplotlib.pyplot as plt

# create x and y data
x = list(range(1, 11))
y = list(range(2, 21, 2))

# generate a line plot
plt.plot(x, y, "r-")

# modify the default y tick marks
plt.yticks([0, 5, 10, 15, 20])

# label the x and y axis 
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

# title the plot
plt.title("A Line Plot")

# show the plot
plt.show()
```
